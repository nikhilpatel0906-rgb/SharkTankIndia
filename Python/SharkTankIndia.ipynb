Shark Tank India Season 1 – Data Analysis
Using Python (Pandas, Matplotlib, Seaborn)

Objective:
Analyze startup pitches, investment behavior, valuation negotiation, and shark investment patterns.

```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Shark Tank India Dataset.csv")

df.head()
```

```python
df.info()
df.describe()
```

```python
total_pitches = df.shape[0]

total_deals = df['deal'].sum()

success_rate = total_deals / total_pitches * 100

total_investment = df['deal_amount'].sum()

avg_deal_amount = df[df['deal']==1]['deal_amount'].mean()

print("Total Pitches:", total_pitches)
print("Total Deals:", total_deals)
print("Success Rate:", round(success_rate,2),"%")
print("Total Investment:", total_investment,"Lakhs")
print("Average Deal:", round(avg_deal_amount,2),"Lakhs")
```

```python
deal_counts = df['deal'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(deal_counts, labels=["Deal","No Deal"], autopct='%1.1f%%')
plt.title("Deal vs No Deal")
plt.show()
```

```python
sharks = {
    "Ashneer": df['ashneer_deal'].sum(),
    "Anupam": df['anupam_deal'].sum(),
    "Aman": df['aman_deal'].sum(),
    "Namita": df['namita_deal'].sum(),
    "Vineeta": df['vineeta_deal'].sum(),
    "Peyush": df['peyush_deal'].sum(),
    "Ghazal": df['ghazal_deal'].sum()
}

shark_df = pd.DataFrame(list(sharks.items()), columns=['Shark','Deals'])

plt.figure(figsize=(8,5))
sns.barplot(x="Shark", y="Deals", data=shark_df)
plt.title("Shark Investment Count")
plt.show()
```

```python
top_funded = df[df['deal']==1].sort_values("deal_amount", ascending=False).head(15)

plt.figure(figsize=(10,6))
sns.barplot(x="deal_amount", y="brand_name", data=top_funded)

plt.title("Top Funded Startups")
plt.xlabel("Investment Amount (Lakhs)")
plt.ylabel("Startup")
plt.show()
```

```python
deal_df = df[df['deal']==1]

plt.figure(figsize=(8,6))
sns.scatterplot(
    x="pitcher_ask_amount",
    y="deal_amount",
    size="deal_equity",
    data=deal_df
)

plt.title("Ask Amount vs Deal Amount")
plt.xlabel("Founder Ask (Lakhs)")
plt.ylabel("Deal Amount (Lakhs)")
plt.show()
```

```python
equity_df = df[['brand_name','ask_equity','deal_equity']][df['deal']==1]

equity_df.plot(
    x='brand_name',
    kind='bar',
    figsize=(12,6)
)

plt.title("Ask Equity vs Deal Equity")
plt.ylabel("Equity %")
plt.show()
```

```python
median_ask_val = df[df['deal']==1]['ask_valuation'].median()

median_deal_val = df[df['deal']==1]['deal_valuation'].median()

negotiation_percent = ((median_deal_val - median_ask_val) / median_ask_val) * 100

print("Median Ask Valuation:", median_ask_val)
print("Median Deal Valuation:", median_deal_val)
print("Negotiation %:", round(negotiation_percent,2))
```

```python
median_ask_equity = df[df['deal']==1]['ask_equity'].median()

median_deal_equity = df[df['deal']==1]['deal_equity'].median()

equity_change = ((median_deal_equity - median_ask_equity) / median_ask_equity) * 100

print("Median Ask Equity:", median_ask_equity)
print("Median Deal Equity:", median_deal_equity)
print("Equity Change %:", round(equity_change,2))
```

Key Findings:

1. 55% of startups secured deals.
2. Total investment in Season 1 was ₹37 Cr.
3. Sharks reduced startup valuation by ~68%.
4. Founders increased equity from 4% to 15% to secure deals.
5. Aman Gupta and Peyush Bansal were the most active investors.
